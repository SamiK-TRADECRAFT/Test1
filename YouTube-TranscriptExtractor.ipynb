```python
import time
from selenium import webdriver

driver = webdriver.Chrome('D:\DOWNLOADS\CHROMEDRIVER\chromedriver.exe')

mcintyre = 'https://www.youtube.com/watch?v=eC6eHraeMes'
timurban = 'https://www.youtube.com/watch?v=arj7oStGLkU'
ualemergency = 'https://www.youtube.com/watch?v=_oq_xEkLRLQ'
spam = 'https://www.youtube.com/watch?v=C4Uc-cztsJo'

snapchat = 'https://www.youtube.com/watch?v=17yiLinMPpk'

driver.get(snapchat)
driver.maximize_window()
time.sleep(2)

button = driver.find_element_by_css_selector("ytd-menu-renderer.ytd-video-primary-info-renderer > yt-icon-button:nth-child(2) > button:nth-child(1)")
button.click()

item = driver.find_element_by_css_selector('ytd-menu-service-item-renderer.style-scope')
item.click()

time.sleep(2)
captions = driver.find_elements_by_css_selector("div.cue-group.style-scope.ytd-transcript-body-renderer")
```

```python
transcript = []

for c in captions:
    clean = c.text
#     print(clean.split('\n'))
    time = clean.split('\n')[0]
#     print(time)
    txt = clean.split('\n')[-1]
    print(txt)
    transcript.append(txt)
#     print('-------------')
transcript = (' ').join(transcript)
```
